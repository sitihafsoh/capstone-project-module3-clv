## **6. Metric Evaluation**

a. Mean Absolute Error (MAE)

- Pengertian: MAE adalah ukuran rata-rata dari selisih absolut antara nilai prediksi dan nilai aktual. Sederhananya, MAE menunjukkan seberapa jauh, rata-rata, prediksi model meleset dari nilai sebenarnya.

- Kelebihan:
  - Mudah diinterpretasikan: MAE memberikan nilai rata-rata kesalahan dalam satuan yang sama dengan variabel target (dalam hal ini, dolar).
  - Tahan terhadap outliers: Outliers tidak memberikan pengaruh yang sangat besar pada nilai MAE karena selisih dihitung secara absolut.

- Hubungan dengan aspek bisnis: MAE membantu perusahaan memahami seberapa besar kesalahan yang mungkin terjadi dalam perkiraan pendapatan dari setiap pelanggan. Semakin kecil MAE, semakin akurat model dalam memprediksi CLV, sehingga perusahaan dapat membuat keputusan yang lebih baik terkait strategi pemasaran, retensi pelanggan, dan alokasi anggaran.

b. Median Absolute Error (MedAE)

- Pengertian: MedAE adalah median dari selisih absolut antara nilai prediksi dan nilai aktual.

- Kelebihan:
  - Lebih tahan terhadap outliers: Sama seperti MAE, MedAE juga tahan terhadap pengaruh outliers. Namun, MedAE lebih robust karena hanya memperhitungkan nilai tengah dari selisih absolut, sehingga tidak terpengaruh oleh nilai ekstrem.

- Hubungan dengan aspek bisnis: MedAE memberikan gambaran tentang kesalahan tipikal dalam prediksi. Nilai MedAE yang rendah menunjukkan bahwa sebagian besar prediksi cukup akurat.

c. Mean Absolute Percentage Error (MAPE)

- Pengertian: MAPE adalah rata-rata dari persentase selisih absolut antara nilai prediksi dan nilai aktual.

- Kelebihan:
  - Mudah diinterpretasikan: MAPE memberikan hasil dalam bentuk persentase, sehingga mudah dibandingkan antar model atau dataset yang berbeda.

- Kelemahan:
  - Sensitif terhadap nilai aktual yang mendekati nol: Jika ada nilai aktual yang sangat kecil (mendekati nol), maka MAPE dapat menghasilkan nilai yang sangat besar atau bahkan tak terdefinisi.

- Hubungan dengan aspek bisnis: MAPE menunjukkan seberapa besar persentase kesalahan rata-rata dalam prediksi. Nilai MAPE yang rendah menunjukkan bahwa model dapat memprediksi CLV dengan akurasi yang tinggi secara relatif terhadap nilai aktualnya.

d. R-squared (Koefisien Determinasi)

- Pengertian: R-squared mengukur proporsi variabilitas dalam variabel dependen (CLV) yang dapat dijelaskan oleh variabel independen (fitur-fitur dalam model).

- Kelebihan:
  - Memberikan gambaran keseluruhan: R-squared memberikan gambaran umum tentang seberapa baik model sesuai dengan data.

- Kelemahan:
  - Tidak sensitif terhadap jenis kesalahan: R-squared tidak membedakan antara overestimation dan underestimation.
  - Dapat dipengaruhi oleh jumlah variabel prediktor: Semakin banyak variabel prediktor, semakin tinggi nilai R-squared, meskipun model mungkin tidak lebih baik dalam memprediksi data baru.

- Hubungan dengan aspek bisnis: R-squared menunjukkan seberapa kuat hubungan antara fitur-fitur dalam model dengan CLV. Nilai R-squared yang tinggi menunjukkan bahwa model dapat menjelaskan sebagian besar variasi dalam CLV.

## **8. Outliers**

```python
# Menampilkan visualisasi outliers dengan boxplot

# Visualisasi outliers dengan boxplot
plt.figure(figsize=(12, 8))
for i, kolom in enumerate(clv.select_dtypes(include=['float', 'int']).columns):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=clv[kolom], color='navy')
    plt.title(f'Boxplot {kolom}')
    plt.ylabel(kolom)
plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah dan persentase outlier
outliers = {}
for kolom in clv.select_dtypes(include=['float', 'int']).columns:
    Q1 = clv[kolom].quantile(0.25)
    Q3 = clv[kolom].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    jumlah_outlier = clv[(clv[kolom] < lower_bound) | (clv[kolom] > upper_bound)][kolom].count()
    persentase_outlier = jumlah_outlier / len(clv) * 100
    outliers[kolom] = {'Jumlah Outlier': jumlah_outlier, 'Persentase Outlier': persentase_outlier}

# Membuat DataFrame dari dictionary outliers
df_outliers = pd.DataFrame.from_dict(outliers, orient='index')

# Menampilkan DataFrame
df_outliers
```

Output di atas menjelaskan bahwa variabel `Number of Policies` memiliki 228 outliers, `Monthly Premium Auto` memiliki 252 outliers, `Total Claim Amount` memiliki 217 outliers, `Income` tidak memiliki outliers, dan `Customer Lifetime Value` memiliki 449 outliers.<br><br>
Pada output boxplot, terlihat ada beberapa box yang memiliki ekstrim outliers. Ekstrim outliers ini dapat menyebabkan hasil analisa statistik dan model prediksi yang menyimpang. Oleh karena itu, outlier-outlier ekstrim ini akan dihapuskan dengan mempertimbangkan menghapus 1/3 outlier terakhir.

**Ekstrim Outliers Variabel `Number of Policies`**

```python
max_num_policies = clv['Number of Policies'].max()
q1_num_policies = clv['Number of Policies'].quantile(0.25)
q3_num_policies = clv['Number of Policies'].quantile(0.75)
iqr_num_policies = q3_num_policies - q1_num_policies
lower_bound_num_policies = q1_num_policies - 1.5 * iqr_num_policies
upper_bound_num_policies = q3_num_policies + 1.5 * iqr_num_policies

# Menampilkan data yang termasuk dalam ekstrim outliers
outliers_num_policies = clv[(clv['Number of Policies'] < lower_bound_num_policies) | (clv['Number of Policies'] > upper_bound_num_policies)]
outliers_num_policies['Number of Policies'].min()

# Menentukan ekstrim outliers
ekstrim_num_policies = max_num_policies - ((max_num_policies - outliers_num_policies['Number of Policies'].min())/3)
ekstrim_num_policies
```

```python
# Hapus baris dengan 'Number of Policies' > ekstrim_num_policies
clv = clv[clv['Number of Policies'] <= ekstrim_num_policies]

# Verifikasi penghapusan
print(clv['Number of Policies'].max())
```

Ouput maksimal outlier setelah penghapusan ekstrim outliers tetap memunculkan nilai 9 artinya, tidak ada ekstrim outliers (1/3 outliers terakhir) pada variabel `Number of Policies`.

## **3. Feature Engineering**

```python
# Membuat pipeline

# Preprocessing fitur-fitur numerikal
num_features = ['Number of Policies', 'Monthly Premium Auto', 'Total Claim Amount', 'Income']
num_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

# Preprocessing fitur-fitur ordinal
ord_features = ['Coverage', 'Education']
categories_ord = [
    ['Basic', 'Extended', 'Premium'], # Basic < Extended < Premium
    ['High School or Below', 'College', 'Bachelor', 'Master', 'Doctor'] # High School or Below < College < Bachelor < Master < Doctor
]
ord_transformer = Pipeline(steps=[
    ('encoder', OrdinalEncoder(categories=categories_ord)),
    ('scaler', RobustScaler())
])

# Preprocessing fitur kategorikal > 5 unique value
cat_features1 = ['Vehicle Class'] # > 5 unique value pakai binary encoder
cat_transformer1 = Pipeline(steps=[
    ('encoder', BinaryEncoder()),
    ('scaler', RobustScaler())
])

# Preprocessing fitur-fitur kategorikal <= 5 unique value
cat_features2 = ['Renew Offer Type', 'Employment Status'] # <=5 unique value pakai one hot encoder
cat_transformer2 = Pipeline(steps=[
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)), # Set sparse_output to False
    ('scaler', RobustScaler(with_centering=False)) # Set with_centering to False
])

# Mengkombinasikan transformer dengan ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('ord', ord_transformer, ord_features),
        ('cat1', cat_transformer1, cat_features1),
        ('cat2', cat_transformer2, cat_features2)
        ], remainder='passthrough')

preprocessor
```

## **3. Hyper Parameter Tuning**

```python
# Hyperparameter Tuning for Random Forest
param_grid_rf = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

pipeline_rf = Pipeline(steps=[('preprocessor', preprocessor), ('model', rf)])

random_search_rf = RandomizedSearchCV(
    pipeline_rf,
    param_distributions=param_grid_rf,
    n_iter=50,
    cv=10,
    scoring='neg_median_absolute_error',
    random_state=42,
    n_jobs=-1
)

random_search_rf.fit(xtrain, ytrain)

best_params_rf = random_search_rf.best_params_
best_score_rf = random_search_rf.best_score_

print("Best Hyperparameters (Random Forest):", best_params_rf)
print("Best MedAE (Random Forest):", -best_score_rf)
```

```python
# Hyperparameter Tuning for Bagging Decision Tree
param_grid_bagging = {
    'model__n_estimators': [10, 50, 100],
    'model__estimator__max_depth': [None, 10, 20],  # Change 'base_estimator__' to 'estimator__'
    'model__estimator__min_samples_split': [2, 5, 10],  # Change 'base_estimator__' to 'estimator__'
    'model__estimator__min_samples_leaf': [1, 2, 4]  # Change 'base_estimator__' to 'estimator__'
}

pipeline_bagging = Pipeline(steps=[('preprocessor', preprocessor), ('model', bagging_dt)])

random_search_bagging = RandomizedSearchCV(
    pipeline_bagging,
    param_distributions=param_grid_bagging,
    n_iter=50,
    cv=10,
    scoring='neg_median_absolute_error',
    random_state=42,
    n_jobs=-1
)

random_search_bagging.fit(xtrain, ytrain)

best_params_bagging = random_search_bagging.best_params_
best_score_bagging = random_search_bagging.best_score_

print("Best Hyperparameters (Bagging Decision Tree):", best_params_bagging)
print("Best MedAE (Bagging Decision Tree):", -best_score_bagging)
```

```python
# Hyperparameter Tuning for Decision Tree
param_grid_dt = {
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10, 20],
    'model__min_samples_leaf': [1, 2, 4, 8]
}

pipeline_dt = Pipeline(steps=[('preprocessor', preprocessor), ('model', dt)])

random_search_dt = RandomizedSearchCV(
    pipeline_dt,
    param_distributions=param_grid_dt,
    n_iter=50,
    cv=10,
    scoring='neg_median_absolute_error',
    random_state=42,
    n_jobs=-1
)

random_search_dt.fit(xtrain, ytrain)

best_params_dt = random_search_dt.best_params_
best_score_dt = random_search_dt.best_score_

print("Best Hyperparameters (Decision Tree):", best_params_dt)
print("Best MedAE (Decision Tree):", -best_score_dt)
```

```python
# Evaluasi model terbaik dengan semua metrik

# Model terbaik setelah tuning
best_rf_tuned = RandomForestRegressor(
    n_estimators=best_params_rf['model__n_estimators'],
    max_depth=best_params_rf['model__max_depth'],
    min_samples_split=best_params_rf['model__min_samples_split'],
    min_samples_leaf=best_params_rf['model__min_samples_leaf'],
    random_state=42
)

best_bagging_tuned = BaggingRegressor(
    estimator=DecisionTreeRegressor(
        max_depth=best_params_bagging['model__estimator__max_depth'],
        min_samples_split=best_params_bagging['model__estimator__min_samples_split'],
        min_samples_leaf=best_params_bagging['model__estimator__min_samples_leaf'],
        random_state=42
    ),
    n_estimators=best_params_bagging['model__n_estimators'],
    random_state=42
)

best_dt_tuned = DecisionTreeRegressor(
    max_depth=best_params_dt['model__max_depth'],
    min_samples_split=best_params_dt['model__min_samples_split'],
    min_samples_leaf=best_params_dt['model__min_samples_leaf'],
    random_state=42
)

# List model terbaik setelah tuning
tuned_models = [best_rf_tuned, best_bagging_tuned, best_dt_tuned]
tuned_model_names = ['Random Forest Tuned', 'Bagging Decision Tree Tuned', 'Decision Tree Tuned']

# Evaluasi model terbaik setelah tuning
tuned_results = {}

for name, model in zip(tuned_model_names, tuned_models):
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
    cv_results = cross_validate(pipeline, xtrain, ytrain, cv=10, scoring=scoring, return_train_score=True)
    tuned_results[name] = cv_results

# Hasil evaluasi model terbaik setelah tuning
tuned_evaluation_results = {}

for model_name, result in tuned_results.items():
    tuned_evaluation_results[model_name] = {
        'mean_test_MAE': np.mean(result['test_MAE']),
        'mean_test_MedAE': np.mean(result['test_MedAE']),
        'mean_test_MAPE': np.mean(result['test_MAPE']),
        'mean_test_R2': np.mean(result['test_R2'])
    }

tuned_eval_df = pd.DataFrame(tuned_evaluation_results).T
tuned_eval_df.sort_values('mean_test_MedAE', ascending=True)
```

Output di atas menjelaskan bahwa setelah dilakukan hyper parameter tuning dari ketiga best model, best model yang terbaik tetap Random Forest dengan nilai metrik MedAE setelah tuning adalah 110,57 (sebelumnya 112,32), nilai metrik MAE -1625,65 (sebelumnya -1623,91), nilai metrik MAPE 11,98% (sebelumnya 11,95%), dan nilai metrik R-squared 67,86% (sebelumnya 67,94%).<br><br>
Jadi best model pada pelatihan ini adalah Random Forest setelah melalui proses tuning dengan best parameternya `model__n_estimators: 100`, `model__min_samples_split: 5`, `model__min_samples_leaf: 4`, dan `model__max_depth: None`.
